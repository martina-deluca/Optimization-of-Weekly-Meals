```python
#Import recipe prices and define c variable for objective function
c = pd.read_csv('recipe_prices.csv', names=['Recipe','Price'],sep=';')
c=c.set_index('Recipe')
c.values
```

```python
c.shape
```

```python
for i in range(0,7):
    #Define contraints
    #Redefine for each day/loop since we are removing recipes
    T = data[data.columns.intersection(['breakfast','lunch','snack','dinner','dessert'])].T
    if (diet_choice!=''):
        f = data[data.columns.intersection([diet_choice])].T
    A = data[data.columns.intersection(['rating','calories','protein','fat','sodium','vegetable','fruit'])].T
    max_rating = 5
    b_max = np.array([max_rating,max_calories,max_fat,max_protein,max_sodium,max_veg,max_fruit])
    d_min = np.array([min_rating,min_calories,min_fat,min_protein,min_sodium,min_veg,min_fruit])
    #Incorporate all inequality constraints in one matrix and vector
    Aineq = pd.concat([A,-A]).to_numpy()
    bineq = np.concatenate([b_max,-d_min])
    #Incorporate all equality constraints in one matrix and vector
    if (diet_choice!=''):
        Aeq = np.vstack((np.ones(data.shape[0]).T,T.to_numpy(),f.to_numpy()))
        beq = [5,1,1,1,1,1,5]
    else:
        Aeq = np.vstack((np.ones(data.shape[0]).T,T.to_numpy()))
        beq = [5,1,1,1,1,1]
    #Define and solve the CVXPY problem
    x = cp.Variable(data.shape[0])
    objective = cp.Minimize(c.values.T@x)
    constraints = [Aineq @ x <= bineq, Aeq @ x == beq, x >= 0]
    prob = cp.Problem(objective,constraints)
    prob.solve(solver=cp.ECOS) 

    #Print results
    print('Day',i+1,':')
    print("\nThe optimal value is", prob.value)
    results = pd.DataFrame(data=x.value, index=data.index, columns=['Choice'])
    results = results.merge(T.T, left_on='title', right_on='title')
    #results = results.merge(f.T, left_on='title', right_on='title')
    results = results.sort_values(by='Choice',ascending=False)[0:5]
    results = results.merge(c.loc[results.index,'Price'], left_on='title', right_on='title')
    price = reduce(operator.add, c.loc[results.index,'Price'])
    print("\nThe actual prince is", price)
    display(results)
    #Remove selected recipies from data set and prices vector
    data.drop(results.index.tolist(), axis=0, inplace=True)
    c.drop(results.index.tolist(), axis=0, inplace=True)
```
